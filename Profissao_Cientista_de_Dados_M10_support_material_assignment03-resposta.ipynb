1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_raw = pd.read_csv("previsao_de_renda.csv").drop(columns='Unnamed: 0')
df = df_raw[['data_ref', 'sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia', 'mau']]
#df.head()

def grafico_barras(df, variavel, eixo):
  tab = pd.crosstab(df['data_ref'],df[variavel])
  tab.div(tab.sum(axis = 1), axis = 0)

  tab.plot.bar(ax=eixo, stacked = True)
  tick_labs = df['data_ref'].map(lambda st: st[0:7]).unique()
  labels = eixo.set_xticklabels(tick_labs, rotation=90)

plt.close("all")
fig, axes = plt.subplots(8, 1)
fig.set_figwidth(10)
fig.set_figheight(50)

grafico_barras(df, "sexo", axes[0])
grafico_barras(df, "posse_de_veiculo", axes[1])
grafico_barras(df, "posse_de_imovel", axes[2])
grafico_barras(df, "tipo_renda", axes[3])
grafico_barras(df, "educacao", axes[4])
grafico_barras(df, "estado_civil", axes[5])
grafico_barras(df, "tipo_residencia", axes[6])
grafico_barras(df, "mau", axes[7])
````

Parece que os valores (secundário, superior completo) da variável educação tiveram variação significativa entre 08-2015 e 09-2015

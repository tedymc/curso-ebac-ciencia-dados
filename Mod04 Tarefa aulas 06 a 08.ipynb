####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), e que devolve os dados da consulta no formato DataFrame Seu código aqui

from enum import Enum

#criei um enum para evitar errar o nome dos valores possíveis para tipo interessado
class TipoInteressado(Enum):
    ESTADO = 'Estado'
    MUNICIPIO = 'Munic%C3%ADpio'

def obtem_df_vacinas(uf, tipo_interessado):
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf.upper()}&tipo_interessado={tipo_interessado.value}'
    data_json = requests.get(url, proxies=proxies).json()
    return pd.DataFrame(data_json['items'])
    
obtem_df_vacinas('sp', TipoInteressado.MUNICIPIO).head()    
    
````

````python
# 2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas

df_vacinas = obtem_df_vacinas('mg', TipoInteressado.ESTADO)
(df_vacinas['status'] == 'Arquivado por decurso de prazo').sum()
````

````python
# 3. Qual é o município da Bahia com mais solicitações deferidas?
df_vacinas = obtem_df_vacinas('ba', TipoInteressado.MUNICIPIO)
df_vacinas[df_vacinas['status'] == 'Deferido']['interessado'].value_counts()[0:1]
````

````python
# 4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
df_vacinas = obtem_df_vacinas('ba', TipoInteressado.ESTADO)
df_vacinas.to_csv('solicitacoes_BA_Estado.csv')
````

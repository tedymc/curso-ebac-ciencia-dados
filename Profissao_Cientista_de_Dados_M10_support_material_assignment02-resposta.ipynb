1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_raw = pd.read_csv("previsao_de_renda.csv").drop(columns='Unnamed: 0')
#df_raw.columns
df = df_raw[['posse_de_imovel', 'posse_de_veiculo', 'renda']]

ax = sns.pointplot(x="posse_de_imovel", y="renda", data=df)
ax.grid(axis='y')
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", data=df)
ax.grid(axis='y')
````

Parece que as duas variáveis descrevem bem a renda para toda a população, já que os intervalos de confiança (erro padrão) de cada um dos valores da variável qualitativa não tem intersecção.

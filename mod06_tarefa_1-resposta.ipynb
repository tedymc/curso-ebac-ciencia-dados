####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10,10))
fig, axes = plt.subplots(2, 2)
fig.tight_layout(pad=5.0)

g1 = sns.countplot(data=df, x='SEXO', ax=axes[0,0])
g1.set(title='Sexo', xlabel=None, ylabel=None)

df.loc[df['GESTACAO']=='37 a 41 semanas', 'GESTACAO_ABREV'] = '37-41'
df.loc[df['GESTACAO']=='42 semanas e mais', 'GESTACAO_ABREV'] = '42+'
df.loc[df['GESTACAO']=='32 a 36 semanas', 'GESTACAO_ABREV'] = '32-36'
df.loc[df['GESTACAO']=='22 a 27 semanas', 'GESTACAO_ABREV'] = '22-27'
df.loc[df['GESTACAO']=='28 a 31 semanas', 'GESTACAO_ABREV'] = '28-31'
df.loc[df['GESTACAO']=='Menos de 22 semanas', 'GESTACAO_ABREV'] = '22-'

g2 = sns.countplot(data=df, x='GESTACAO_ABREV', ax=axes[0,1], order=['22-','22-27','28-31','32-36','37-41','42+'])
g2.set(title='Semanas de Gestação', xlabel=None, ylabel=None)

g3 = sns.countplot(data=df, x='ESTCIVMAE', ax=axes[1,0])
g3.set(title='Estado Civil da Mãe', xlabel=None, ylabel=None)
g3.set_xticklabels(['Solteira', 'Casada', 'Viúva', 'Divorciada', '???', 'Ignorado'], rotation=30, fontsize='small')

g4 = sns.countplot(data=df, x='CONSULTAS', ax=axes[1,1])
g4.set(title='Consultas Pré-Natal', xlabel=None, ylabel=None)
g4.set_xticklabels(['Nenhuma', '1-3', '4-6', '7+', 'Ignorado'], rotation=30, fontsize='small')

```

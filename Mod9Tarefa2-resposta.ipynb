### 1. Crie 2 faixas de Latitude do município (munResLat) sendo uma acima e outra abaixo de -10.5 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
sinasc_raw.info()
```

```python
sinasc_raw.loc[sinasc_raw['munResLat'] >= -10.5, 'LAT_FAIXA'] = 'acima'
sinasc_raw.loc[sinasc_raw['munResLat'] < -10.5, 'LAT_FAIXA'] = 'abaixo'

aggr_funcs = [('soma', 'sum'), ('média','mean'), ('mínimo','min'),('máximo','max'), ('mediana','median'), ('desvio padrão','std'), ('variância', 'var')]
aggr_dict = {'IDADEMAE': aggr_funcs, 'IDADEPAI': aggr_funcs}
sinasc_raw.groupby('LAT_FAIXA', dropna=False).aggregate(aggr_dict)
```

### 2. Crie 2 faixas da área dos municípios (munResArea) sendo uma acima e outra abaixo de 3000 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas


```python
sinasc_raw.loc[sinasc_raw['munResArea'] >= 3000, 'AREA_FAIXA'] = 'acima'
sinasc_raw.loc[sinasc_raw['munResArea'] < 3000, 'AREA_FAIXA'] = 'abaixo'

aggr_funcs = [('soma', 'sum'), ('média','mean'), ('mínimo','min'),('máximo','max'), ('mediana','median'), ('desvio padrão','std'), ('variância', 'var')]
aggr_dict = {'QTDFILVIVO': aggr_funcs, 'QTDFILMORT': aggr_funcs}
sinasc_raw.groupby('AREA_FAIXA', dropna=False).aggregate(aggr_dict)
```

### 3. Determine faixas na variável munResAlt e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

```python
sinasc_raw['munResAlt'].describe()
```

```python
sinasc_raw.loc[sinasc_raw['munResAlt'] <= 87.00, 'munResAlt_Faixa'] = 'faixa1'
sinasc_raw.loc[(sinasc_raw['munResAlt'] > 87.00)  & (sinasc_raw['munResAlt'] <= 153.00), 'munResAlt_Faixa'] = 'faixa2'
sinasc_raw.loc[(sinasc_raw['munResAlt'] > 153.00) & (sinasc_raw['munResAlt'] <= 191.00), 'munResAlt_Faixa'] = 'faixa3'
sinasc_raw.loc[sinasc_raw['munResAlt'] > 191.00, 'munResAlt_Faixa'] = 'faixa4'

aggr_funcs = [('soma', 'sum'), ('média','mean'), ('mínimo','min'),('máximo','max'), ('mediana','median'), ('desvio padrão','std'), ('variância', 'var')]
aggr_dict = {'QTDPARTNOR': aggr_funcs, 'QTDPARTCES': aggr_funcs}
sinasc_raw.groupby('munResAlt_Faixa', dropna=False).aggregate(aggr_dict)
```
